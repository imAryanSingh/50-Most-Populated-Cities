```python
"""
FITTING the MODEL (Dataset Inputs with the outputs) using LinearRegression.fit(x,y)
We are plotting the values of x and y in the LINEAR REGRESSION LIBRARY
"""
x=dataset[['pop2010']].values
y=dataset[['pop2022']].values
model =LinearRegression()
model.fit(x,y)
```

```python
#22)Using the Input values(pop2010,pop2022), We predict the output value's of Population 2022 for diffrent cities
pop2022_predicted = model.predict(x)
print(pop2022_predicted)
```

```python
#23)This Is the Actual Values of Population in 2022
print(dataset['pop2022'])
"""
When we compare pop2022_pred and pop2022,know there is huge diffrence in there corresponding values
it only means that our model is not linear
"""
```

```python
#24)INDIVIDUAL PREDICTION now for 9000000 Population,I want to know the population of 2010
model.predict([[9000000]])
```

```python
"""
Cross Verification Technique
y=mx+c-Equation of a straight line
m - slope
C - Constant/y - intercept
y - dependent variable
x - Inpendent variable
"""
#25)Slope(m) of model
m=model.coef_
m
```

```python
#26)y-intercept(c) of model
C = model.intercept_
C
```

```python
#27)Now Substituting the value of m and C in y=mx+c
m*9000000+C
#THIS MEANS OUR PREDICTION WAS RIGHT IN STEP - 24
```

```python
#28)Visualization for the best fit line
plt.scatter(x,y)
plt.plot(x,pop2022_predicted,c='yellow')
plt.title('BEST FIT LINE')
```
